### 统计各GraphLabel都包含哪些样本，将 EnrichmentMatrix 按 label 归类

```python
import numpy as np
import pandas as pd
import os

InputFolder = "../../SKCM_Input/"
Region_filename = InputFolder + "ImageNameList.txt"
region_name_list = pd.read_csv(
        Region_filename,
        sep="\t",  # tab-separated
        header=None,  # no heading row
        names=["Image"],  # set our own names for the columns
    )

A_0=[]
B_1=[]
#C_2=[]

for graph_index in range(0, len(region_name_list)):
    
    region_name = region_name_list.Image[graph_index]
    GraphLabel_filename = InputFolder + region_name + "_GraphLabel.txt"
    graph_label = np.loadtxt(GraphLabel_filename, dtype='int64', delimiter="\t")
    if graph_label == 0: A_0.append(region_name)
    if graph_label == 1: B_1.append(region_name)
    #if graph_label == 2: C_2.append(region_name)

print(A_0,len(A_0))
print(B_1,len(B_1))
#print(C_2,len(C_2))


TargetFolder = "../../data/EnrichScoreMatrix/"

region_list = pd.read_csv(
    os.path.join(InputFolder, "ImageNameList.txt"),
    sep="\t",
    header=None,
    names=["Image"]
)["Image"].tolist()

for region_name in region_list:
    label_path = os.path.join(InputFolder, f"{region_name}_GraphLabel.txt")
    graph_label = np.loadtxt(label_path, dtype="int64", delimiter="\t")

    old_csv = os.path.join(TargetFolder, f"{region_name}_EnrichScoreMatrix.csv")
    if not os.path.isfile(old_csv):
        continue

    new_filename = f"{graph_label}_{region_name}_EnrichScoreMatrix.csv"
    new_csv = os.path.join(TargetFolder, new_filename)
    os.rename(old_csv, new_csv)
```
